### Distribuci√≥n aleatoria uniforme sobre circulo y esfera

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(1000)


R=1
theta = (360*x)
r = (2*x)**(1/2)


fig = plt.figure()
ax = fig.add_subplot(projection='polar')
ax.set_rlim(0,2)
c = ax.scatter(theta, r)
```

Ahora con la esfera


```python
plt.scatter(r*np.cos(theta), r*np.sin(theta))
```
