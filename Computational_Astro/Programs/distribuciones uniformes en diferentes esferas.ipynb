### Distribuci√≥n aleatoria uniforme sobre circulo y esfera

If the distribution is uniform
the number of objects per
unit volum is constant:

In sphere is $$dn = r^{2} sin(\theta) d\theta dr d\phi$$

```python
import numpy as np
import matplotlib.pyplot as plt

NumberRands = 1000
rand1 = np.random.rand(NumberRands)
rand2 = np.random.rand(NumberRands)

R=1
theta = (360*rand1)
r = (rand2)**(1/2)


fig = plt.figure()
ax = fig.add_subplot(projection='polar')
ax.set_rlim(0,2)
ax.scatter(theta, r)
```

Ahora con la esfera


```python
ang = np.linspace(0,2*np.pi,100)
plt.scatter(r*np.cos(theta), r*np.sin(theta))
plt.plot(np.cos(ang),np.sin(ang), color= "red")
plt.axis('equal')
```

```python
# 

N= 500
rand3 = np.random.rand(N)
rand4 = np.random.rand(N)
rand5 = np.random.rand(N)

R=2
phi = 2*180*rand3
theta = np.arccos(2*rand4-1)
r = R*np.cbrt(rand5)

x = r*np.sin(theta)*np.cos(phi)
y = r*np.sin(theta)*np.sin(phi)
z = r*np.cos(theta)

fig1 = plt.figure()
ax = fig1.add_subplot(projection='3d')
ax.scatter(x,y,z, s=5)
plt.show()

# Print two graphs
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(9, 4), sharey=True)


ax1.scatter(x,y)
ax1.set_xlim(-1.5,1.5)
ax1.set_ylim(-1.5,1.5)
ax1.plot(R*np.cos(ang),R*np.sin(ang), color= "red")
ax1.axis('equal')


ax2.scatter(y,z)
ax2.set_xlim(-1.5,1.5)
ax2.set_ylim(-1.5,1.5)
ax2.plot(R*np.cos(ang),R*np.sin(ang), color= "red")
ax2.axis('equal')
plt.show()
```

```python
#Rejection method
N = 10000
rand6  = np.random.rand(N)
rand7 = np.random.rand(N)
rand8 = np.random.rand(N)

R = 1
x = 2*R*rand6 - R
y = 2*R*rand7 - R
z = 2*R*rand8 - R

x_g = []
y_g = []
z_g = []

for i in range(N):
    if x[i]**2 +y[i]**2+z[i]**2 < R**2:
        x_g.append(x[i])
        y_g.append(y[i])
        z_g.append(z[i])
    else:
        None


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x_g,y_g,z_g, s=5)
plt.show()
        
```
